```python
# 세 정수를 입력받아 최댓값 구하기

print('세 정수의 최댓값을 구합니다')    
a = input('정수 a의 값을 입력하세요')
b = input('정수 b의 값을 입력하세요')
c = input('정수 c의 값을 입력하세요')

max = a
#조건식으로 평가한 결과에 따라 프로그램의 실행 흐름이 변경되는데 이러한 구조를 선태구조라 한다
if b > max : 
    max = b
if c > max : 
    max = c

print(f'최댓값은 {max}입니다')
```

```python
# 세 정수의 최댓값 구하기
def max3(a, b, c):
    #a, b, c의 최댓값을 구하여 반환
    max = a
    if b > max :
        max = b
    if c > max :
        max = c
    return max

print(f'max3(3, 2, 1) = {max3(3, 2, 1)}') # a > b > c
print(f'max3(3, 2, 2) = {max3(3, 2, 2)}') # a > b = c
print(f'max3(3, 1, 2) = {max3(3, 1, 2)}') # a > c > b
print(f'max3(3, 2, 3) = {max3(3, 2, 3)}') # a = c > b
print(f'max3(3, 3, 3) = {max3(3, 3, 3)}') # a = b = c

#알고리즘: 어떠한 문제를 해결하기 위해 정해 놓은 일련의 절차
```

```python
# 세 정수를 입력받아 중앙값을 구하기
def med3(a, b, c) :
    '''a, b, c의 중앙값을 구하여 반환'''
    if a >= b :
        if b >= c :
            return b
        elif a <= c :
            return a
        else :
            return c
    elif a > c : 
        return a
    elif b > c :
        return c
    else :
        return b

print('세 정수의 중앙값을 구합니다')
a = input('정수 a의 값을 입력하세요')
b = input('정수 b의 값을 입력하세요')
c = input('정수 c의 값을 입력하세요')

print(f'중앙값은 {med3(a, b, c)}입니다')
```

```python
# 입력받은 정수의 부호(양수, 음수, 0)출력하기

n = int(input('정수를 입력하세요: '))

if n > 0 :
    print('이 수는 양수입니다')
elif n < 0 :
    print('이 수는 음수입니다')
else :
    print('이 수는 0입니다')

```

```python
# 3개로 분기하는 조건문
# n이 1이면 A, 2이면 B, 그 외(1도 2도 아닌 모든 값)에는 C를 출력
# else는 위의 모든 조건이 거짓일 때 실행되는 기본분기

n = int(input('정수를 입력하세요: '))

if n == 1 :
    print('A')
elif n == 2 :
    print('B')
else :
    print('C')
```

```python
# 4개로 분기하는 조건문
# n이 1이면 A, 2이면 B, 3이면 C를 출력
# 그 외의 값이면 pass(아무것도 하지 않음)로 처리
# elif를 추가하면 분기를 원하는 만큼 늘릴 수 있다

n = int(input('정수를 입력하세요: '))

if n == 1 :
    print('A')
elif n == 2 :
    print('B')
elif n == 3 :
    print('C')
else :
    pass
```
