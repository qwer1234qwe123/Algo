```python
# 세 정수의 최댓값 구하기
def max3(a, b, c):
    #a, b, c의 최댓값을 구하여 반환
    max = a
    if b > max :
        max = b
    if c > max :
        max = c
    return max

print(f'max3(3, 2, 1) = {max3(3, 2, 1)}') # a > b > c
print(f'max3(3, 2, 2) = {max3(3, 2, 2)}') # a > b = c
print(f'max3(3, 1, 2) = {max3(3, 1, 2)}') # a > c > b
print(f'max3(3, 2, 3) = {max3(3, 2, 3)}') # a = c > b
print(f'max3(3, 3, 3) = {max3(3, 3, 3)}') # a = b = c

#알고리즘: 어떠한 문제를 해결하기 위해 정해 놓은 일련의 절차
```

```python
# 1부터 n까지 정수의 합 구하기(for문)
# range() 함수는 이터러블 객체를 생성할 수 있음
# 이터러블 객체는 반복할 수 있는 객체를 말하며 파이썬에서 사용하는 대표적인 이터러블 자료형으로 list, str, tuple이 있음
# range() 함수로 이터러블 객체를 생성하는 방법
#   1. for i in range(n): 0 이상 n 미만인 수를 차레로 나열하는 수열
#   2. for i in range(a, b): a 이상 b 미만인 수를 차례로 나열하는 수열
#   3. for i in range(a, b, step): a 이상 b 미만인 수를 step 간격으로 나열하는 수열

print('1부터 n까지의 정수의 합을 구합니다')
n = int(input('n값을 입력하세요'))

sum = 0
for i in range (1, n + 1) :
    sum += i

print(f'1부터 {n}까지의 합은 {sum}입니다')
```

```python
# 가우스 덧셈으로 1부터 n까지 정수의 합 구하기
# 가우스 덧셈 수학식: n(n+1) / 2
print('1부터 n까지의 정수의 합을 구합니다')
n = int(input('n값을 입력하세요'))

sum = n * (n + 1) // 2

print(f'1부터 {n}까지의 합은 {sum}입니다')
```

```python
# a부터 b까지의 정수의 합 구하기(for문)
print('a부터 b까지의 정수의 합을 구합니다')
a = int(input('a의 값을 입력'))
b = int(input('b의 값을 입력'))

if a > b :
    a, b = b, a # a와 b를 오름차순으로 정렬 (a와 b의 값을 교환)

sum = 0

for i in range(a, b + 1):
    sum += i

print(f'{a}부터 {b}까지의 합은 {sum}입니다')
```

```python
# a부터 b까지의 정수의 합 구하기 1

print('a부터 b까지의 정수의 합을 구합니다')
a = int(input('정수 a를 입력하세요'))
b = int(input('정수 b를 입력하세요'))

if a > b :
    a, b = b, a

sum = 0
for i in range(a, b + 1) :
    if i < b :          # i가 b보다 작으면 합을 구하는 과정 출력
        print(f'{i} + ', end='')
    else :              # i가 b보다 크거나 같으면 최종값 출력을 위해 i = 를 출력
        print(f'{i} = ', end='')
    sum += i            # sum에 i를 더함

print(sum)
```

```python
# +와 -를 번갈아 출력하기 2
print('+와 -를 번갈아 출력합니다')
n = int(input('몇개를 출력할까요?'))

# _(언더스코어): 반복 변수의 값을 사용하지 않을 때 쓰는 관례적 표시
# 여기서는 단지 n//2번 반복만 하면 되고, 변수 값은 필요 없으므로 _ 사용
# n // 2 여기서 (//)은 정수 나눗셈(몫만 구하기) 연산자
for _ in range(n // 2): 
    print('+-', end='') #+-를 n//2개의 출력

if n % 2:
    print('+', end='')

print()
```

```python
# +를 n개 출력하되 w개 마다 줄바꿈하기 1

print('*를 출력합니다')
n = int(input('몇 개를 출력할까요'))
w = int(input('몇 개마다 줄바꿈할까요'))

for i in range(n) :         # n번 반복
    print('*', end='')      
    if i % w == w - 1 :     # n번 판단
        print()             # 줄바꿈
    
if n % w :                  # 1번 판단
    print()                 # 줄바꿈
```

```python
# +를 n개 출력하되 w개 마다 줄바꿈하기 2

print('*를 출력합니다')
n = int(input('몇 개를 출력할까요'))
w = int(input('몇 개마다 줄바꿈할까요'))

for _ in range(n // w) :    # n // w번 반복
    print('*' * w)
    
rest = n % w
if rest:                    # if문 판단 1번
    print('*' * rest)       
```

```python
# 1부터 n까지의 정수의 합 구하기(n값은 양수만 입력 받음)

print('1부터 n까지의 정수의 합을 구합니다')

while True : 
    n = int(input('n값을 입력하세요'))
    if n > 0 :
        break   # n이 0보다 커질 때까지 반복

sum = 0
i = 1

for i in range(1, n + 1) :
    sum += i    # sum에 i를 더함
    i += 1      # i에 1을 더함

print(f'1부터 {n}까지의 정수의 합은 {sum}입니다')
```

```python
# 가로, 세로 길이가 정수이고 넓이가 area인 직사각형에서 변의 길이 나열하기

area = int(input('직사각형의 넓이를 입력하세요'))

for i in range(1, area + 1):
    if i * i > area: 
        break
    if area % i :
        continue
    print(f'{i} x {area // i}')
```

```python
# 왼쪽 아래가 직각인 이등변 삼각형으로 * 출력하기

print('왼쪽 아래가 직각인 이등변 삼각형을 출력합니다')
n = int(input('짧은 변의 길이를 입력하세요'))


for i in range(n):
    for j in range(i + 1):
        print('*', end='')
    print()
```

```python
# 오른쪽 아래가 직각인 이등변 삼각형으로 + 출력하기

print('오른쪽 아래가 직각인 이등변 삼각형을 출력합니다.')

n = int(input('짧은 변의 길이를 입력하세요'))

for i in range(n):
    for _ in range(n - i -1):
        print(' ', end='')
    for _ in range(i + 1):
        print('*', end='')
    print()
```
