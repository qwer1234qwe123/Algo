```python
# 세 정수의 최댓값 구하기
def max3(a, b, c):
    #a, b, c의 최댓값을 구하여 반환
    max = a
    if b > max :
        max = b
    if c > max :
        max = c
    return max

print(f'max3(3, 2, 1) = {max3(3, 2, 1)}') # a > b > c
print(f'max3(3, 2, 2) = {max3(3, 2, 2)}') # a > b = c
print(f'max3(3, 1, 2) = {max3(3, 1, 2)}') # a > c > b
print(f'max3(3, 2, 3) = {max3(3, 2, 3)}') # a = c > b
print(f'max3(3, 3, 3) = {max3(3, 3, 3)}') # a = b = c

#알고리즘: 어떠한 문제를 해결하기 위해 정해 놓은 일련의 절차
```

```python
# 1부터 n까지 정수의 합 구하기(while문)
print('1부터 n까지의 정수의 합을 구합니다')
n = int(input('n값을 입력하세요'))

sum = 0

while i <= n :
    sum += i
    i += 1

int(f'1부터 {n}까지의 합은 {sum}입니다')
```

```python
# 1부터 n까지 정수의 합 구하기(for문)
# range() 함수는 이터러블 객체를 생성할 수 있음
# 이터러블 객체는 반복할 수 있는 객체를 말하며 파이썬에서 사용하는 대표적인 이터러블 자료형으로 list, str, tuple이 있음
# range() 함수로 이터러블 객체를 생성하는 방법
#   1. for i in range(n): 0 이상 n 미만인 수를 차레로 나열하는 수열
#   2. for i in range(a, b): a 이상 b 미만인 수를 차례로 나열하는 수열
#   3. for i in range(a, b, step): a 이상 b 미만인 수를 step 간격으로 나열하는 수열

print('1부터 n까지의 정수의 합을 구합니다')
n = int(input('n값을 입력하세요'))

sum = 0
for i in range (1, n + 1) :
    sum += i

print(f'1부터 {n}까지의 합은 {sum}입니다')
```

```python
# 가우스 덧셈으로 1부터 n까지 정수의 합 구하기
# 가우스 덧셈 수학식: n(n+1) / 2
print('1부터 n까지의 정수의 합을 구합니다')
n = int(input('n값을 입력하세요'))

sum = n * (n + 1) // 2

print(f'1부터 {n}까지의 합은 {sum}입니다')
```

```python
# a부터 b까지의 정수의 합 구하기(for문)
print('a부터 b까지의 정수의 합을 구합니다')
a = int(input('a의 값을 입력'))
b = int(input('b의 값을 입력'))

if a > b :
    a, b = b, a # a와 b를 오름차순으로 정렬 (a와 b의 값을 교환)

sum = 0

for i in range(a, b + 1):
    sum += i

print(f'{a}부터 {b}까지의 합은 {sum}입니다')
```
